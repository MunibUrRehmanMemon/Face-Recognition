# Convert Images to CSV

```python

model = insightface.app.FaceAnalysis()
model.prepare(ctx_id=0)

# Loc of labeled images (each folder represents a different person)
labeled_faces_dir = '/path/to/lebeled/images'

def save_embeddings_to_csv(labeled_faces_dir, output_csv):
    embeddings_data = []
    for foldername in os.listdir(labeled_faces_dir):
        folder_path = os.path.join(labeled_faces_dir, foldername)
        if os.path.isdir(folder_path):
            for filename in os.listdir(folder_path):
                if filename.endswith(('.jpg', '.jpeg', '.png')):
                    image_path = os.path.join(folder_path, filename)
                    image = cv2.imread(image_path)
                    rgb_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
                    if rgb_img is None:
                        print(f"Error: Could not read image {image_path}")
                        continue
                    faces = model.get(rgb_img)
                    if faces:
                        face = faces[0]
                        embeddings_data.append({
                            'label': foldername,
                            'filename': filename,
                            'embedding': face.embedding.tolist()
                        })
                        print(f"Processed {foldername}/{filename}")

    # Save as CSV
    df = pd.DataFrame(embeddings_data)
    df.to_csv(output_csv, index=False)
    print(f"Embeddings saved to {output_csv}")

output_csv = '/path/to/output/dataset.csv'
save_embeddings_to_csv(labeled_faces_dir, output_csv)
```

# Face Recognition through Video

```python


# Function to Load Embeddings from dataset.csv
def load_embeddings_from_csv(csv_file):
    labeled_faces = {}
    try:
        df = pd.read_csv(csv_file)

        for _, row in df.iterrows():
            label = row['label']
            embedding = np.array(row['embedding'][1:-1].split(', '), dtype=float)

            if label in labeled_faces:
                labeled_faces[label].append(embedding)
            else:
                labeled_faces[label] = [embedding]

        # Compute the average
        for label, embeddings in labeled_faces.items():
            labeled_faces[label] = np.mean(embeddings, axis=0)
    except Exception as e:
        print(f"Error loading CSV embeddings: {str(e)}")
    return labeled_faces
```

# Encode Faces from a New Video

```python

def encode_faces_from_new_video(video_path, output_pkl_file):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    frame_count = 0
    encoded_faces = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Convert frame to RGB and detect faces
        rgb_img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        faces = model.get(rgb_img)
        frame_encodings = []

        # Extract embeddings
        for face in faces:
            encoding = {
                'bbox': face.bbox.astype(int),
                'embedding': face.embedding
            }
            frame_encodings.append(encoding)

        if frame_encodings:
            encoded_faces.append(frame_encodings)

        frame_count += 1
        if frame_count % 100 == 0:
            print(f"Processed {frame_count} frames")

    cap.release()
    print("Encoding completed.")

    # Save encoded faces to pickle file
    with open(output_pkl_file, 'wb') as f:
        pickle.dump(encoded_faces, f)
    print(f"Encoded faces saved to {output_pkl_file}")
```

# Function to Match Faces and Generate Video

```python

def match_faces_and_generate_video(video_path, output_path, encoded_faces, csv_file):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Load labeled embeddings from CSV
    labeled_faces = load_embeddings_from_csv(csv_file)

    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    frame_count = 0
    for frame_encodings in encoded_faces:
        ret, frame = cap.read()
        if not ret:
            break

        for encoding in frame_encodings:
            bbox = encoding['bbox']
            embedding = encoding['embedding'] / np.linalg.norm(encoding['embedding'])

            best_match_label = "Unknown"
            best_match_distance = float('inf')

            for label, labeled_embedding in labeled_faces.items():
                labeled_embedding = labeled_embedding / np.linalg.norm(labeled_embedding)
                distance = np.linalg.norm(embedding - labeled_embedding)

                if distance < best_match_distance:
                    best_match_distance = distance
                    best_match_label = label

            # Annotating video frame
            if best_match_distance < 1.0:
                cv2.putText(frame, f'{best_match_label} ({best_match_distance:.2f})',
                            (bbox[0], bbox[1] - 10), cv2.FONT_HERSHEY_SIMPLEX,
                            0.5, (0, 255, 0), 2, cv2.LINE_AA)
            else:
                cv2.putText(frame, f'Unknown ({best_match_distance:.2f})',
                            (bbox[0], bbox[1] - 10), cv2.FONT_HERSHEY_SIMPLEX,
                            0.5, (0, 255, 0), 2, cv2.LINE_AA)

            cv2.rectangle(frame, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (0, 255, 0), 2)

        out.write(frame)
        frame_count += 1
        if frame_count % 100 == 0:
            print(f"Processed {frame_count} frames")

    cap.release()
    out.release()
    cv2.destroyAllWindows()
    print("Video generation completed.")
```
